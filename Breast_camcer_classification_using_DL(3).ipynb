**Importing** **the** **dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

**Load_the** **data**

```python
breast_data=sklearn.datasets.load_breast_cancer()
```

```python
breast_data
```

```python
data_frame=pd.DataFrame(breast_data.data,columns=breast_data.feature_names)
```

```python
data_frame.head()
```

**Add target label**

```python
data_frame['label']=breast_data.target
```

```python
data_frame.tail()
```

```python
# shape of the data
data_frame.shape
```

```python
# copy of the data
data=data_frame.copy()
```

```python
data.head()
```

```python
# info of the data
data.info()
```

```python
data.isnull().sum()
```

```python
# describe of the data
data.describe()
```

```python
# split the data
X=data.drop(columns="label")
Y=data["label"]
```

```python
print(X)
print(Y)
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
# apply standardization data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(x_train)
X_test_scaled=scaler.transform(x_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

**Build Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from  tensorflow import keras
```

```python
# build model
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(2,activation="sigmoid")
])
```

```python
#  compile the model
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
# summary of the model
model.summary()
```

```python
history=model.fit(X_train_scaled,y_train,validation_split=0.1,epochs=10)
```

**Visualization the loss and accuracy**

```python
# visulize the accuracy
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("accuracy of train and test")
plt.xlabel("epochs")
plt.ylabel("accuracy")

plt.legend(["train","test"],loc="upper left")
plt.show()
```

```python
# visulize the loss
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("loss of train and test")
plt.xlabel("epochs")
plt.ylabel("loss")

plt.legend(["train","test"],loc="upper left")
plt.show()
```

```python
y_pred = model.predict(X_test_scaled[:1])  # Use slicing to pass a batch of size 1
print(y_pred)
```

**Build a preductive system**

```python
input_data=(-0.04462793, -1.41612656, -0.05903514, -0.16234067,  2.0202457 ,
       -0.11323672,  0.18500609,  0.47102419,  0.63336386,  0.26335737,
        0.53209124,  2.62763999,  0.62351167,  0.11405261,  1.01246781,
        0.41126289,  0.63848593,  2.88971815, -0.41675911,  0.74270853,
       -0.32983699, -1.67435595, -0.36854552, -0.38767294,  0.32655007,
       -0.74858917, -0.54689089, -0.18278004, -1.23064515, -0.6268286)

# /change the input_data to numpy array
input_data_as_array=np.asarray(input_data)

# resize the data
input_data_res=input_data_as_array.reshape(1,-1)

# apply scaler
input_data_scaled=scaler.transform(input_data_res)

# prediction
prediction=model.predict(input_data_scaled)
print(prediction)

# apply argmax function
pred_label=[np.argmax(prediction)]
print(pred_label)

if pred_label[0]==0:
  print("the tumor is malignant")

else:
  print("the tumor is benign")
```
